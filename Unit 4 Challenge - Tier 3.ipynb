#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# transpose properties and reset the index
properties = properties.transpose()
properties = properties.reset_index()

# set column names to equal the values in the first row, subsets df to remove first row
properties.columns = properties.iloc[0]
properties = properties.iloc[1:]
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# rename columns
properties = properties.rename({'Unnamed: 0': 'London Boroughs', pd.NaT: 'ID'}, axis=1)
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# melt properties, using 'London Boroughs' and 'ID' as identifier variables, name melted df properties_m
properties_m = pd.melt(properties, id_vars=['London Boroughs', 'ID'], var_name='Month', value_name='Average Price') 
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# check datatypes of properties_m
properties_m.dtypes
```

```python
# convert and recheck datatypes of properties_m
convert_dict = {'Average Price':'float64'}
properties_m = properties_m.astype(convert_dict)
properties_m.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# A list of non-boroughs. 
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']

# check count of properties_m, as well as count of NaNs
properties_m.count()
properties_m.isna().sum()

# subset properties_m to remove non-boroughs
properties_m = properties_m[~properties_m['London Boroughs'].isin(nonBoroughs)]
properties_m.dropna(inplace=True)

# check new count of properties_m
properties_m.count()

# format string for borough name to camel case 
properties_m['London Boroughs'] = properties_m['London Boroughs'].str.title()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# subset on 'City of London' borough and plot 
city_of_london = properties_m[properties_m['London Boroughs'] == 'City Of London']
city_of_london_plot = city_of_london.plot(x='Month', y='Average Price', title='Average House Price for London Borough: City of London')

#format y-axis with dollar sign 
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
city_of_london_plot.yaxis.set_major_formatter(tick) 

# add labels to plot
plt.xlabel('Month')
plt.ylabel('Average House Price')
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
# extract year from Month value, store in new column 'Year'
properties_m['Year'] = properties_m['Month'].apply(lambda x: x.year)
```

```python
# create copy of properties_m and drop columns 'Month' and 'ID'
properties_m_copy = properties_m
properties_m_copy = properties_m_copy.drop(columns=['Month', 'ID'])

# using copy of properties_m group by 'Year' then 'London Boroughs' to get the mean of the last column, Average House Price, then reset index
properties_m_c = properties_m_copy.groupby(['Year', 'London Boroughs']).mean().reset_index()
```

```python

### plot Average House Prices by Year and London Borough ###

fig, ax = plt.subplots(figsize=(20,20))

# loop through each borough to plot and add label to legend
for name, group in properties_m_c.groupby('London Boroughs'):
    group.plot(x='Year', y='Average Price', ax=ax, label=name)


# format y-axis with dollar sign 
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick) 

# add legend and move to side
ax.legend(loc='upper left', title='London Boroughs', bbox_to_anchor=(1.05, 1))


# add labels, title to plot
plt.xlabel('Year')
plt.ylabel('Average House Price')
plt.title('Average House Prices for London Boroughs')
plt.show()
```

```python
# noting the graph, I sought to identify which borough had the highest average house price consistently from 1998 to 2018
index_max_avg_price = properties_m_c['Average Price'].argmax()
assumed_max = properties_m_c.iloc[index_max_avg_price]
assumed_max_borough = assumed_max['London Boroughs']
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# get list of all 32 London Boroughs
boroughs = properties_m_c['London Boroughs'].unique().tolist()

# function to create price ratio for each borough from 1998 to 2018
def create_price_ratio(borough_name):
    avg_price_1998 = properties_m_c.loc[(properties_m_c['Year'] == 1998) & (properties_m_c['London Boroughs'] == borough_name)]['Average Price']
    avg_price_1998 = avg_price_1998.values[0]
    avg_price_2018 = properties_m_c.loc[(properties_m_c['Year'] == 2018) & (properties_m_c['London Boroughs'] == borough_name)]['Average Price']
    avg_price_2018 = avg_price_2018.values[0]
    
    price_ratio = round((avg_price_1998 / avg_price_2018), 3)
    final = [borough_name, price_ratio]
    
    return final


# maps create_price_ratio function to list of London Boroughs, creates dataframe and sorts by price ratio
all_price_ratios = list(map(create_price_ratio, boroughs))
df = pd.DataFrame(all_price_ratios, columns = ['London Borough', 'Price Ratio']).sort_values('Price Ratio', ascending=False).reset_index(drop=True)


# defines a list of and prints top 5 boroughs with greatest Price Ratio
max_avg_increase_borough2 = list(df['London Borough'].head(5))
max_avg_increase_top5 = df.iloc[:5]
print(max_avg_increase_top5)


# this is the London Borough with the greatest Price Ratio
max_avg_increase_borough = df['London Borough'].iloc[0]


# subsets and plot properties_m_c for London Boroughs in top 5 boroughs with greatest Price Ratio
max_boroughs = properties_m_c[properties_m_c['London Boroughs'].isin(max_avg_increase_borough2)]

fig, ax2 = plt.subplots(figsize=(10,10))
for name, group in max_boroughs.groupby('London Boroughs'):
    group.plot(x='Year', y='Average Price', ax=ax2, label=name)

#format y-axis with dollar sign 
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax2.yaxis.set_major_formatter(tick) 

# add labels, title to plot
plt.xlabel('Year')
plt.ylabel('Average House Price')
plt.title('Average House Prices of Top 5 London Boroughs with Greatest Price Ratio from 1998 to 2018')
plt.show()

```
